У нас **14%** сотрудников имеют стаж отрицательный, я сначала *заменила на 0*, но поняла, что 14% - значимая доля, поэтому **заменяю на медиану по сотрудникам такого возраста**

Первым делом создаю **колонку-флаг**, чтобы запомнить где у нас были некорректные данные по стажу (пропуски и отриц значения)

А затем заменяю пропуски по стажу глобальной медианой по корректным данным  после удаления отрицательного стажа.

Также я проверила колонки на выбросы: у нас максимальный возраст сотрудников 80, у летних месяцев есть только (6,7,8, NaN), нет кол-ва детей меньше 0, а north_bonus_days хранит константу 24 дня или 0.
ВСЕ ОК!

```python
#Создаю флаг ДО любых замен
df['tenure_was_invalid'] = (
    (df['tenure_months'] < 0) |
    (df['tenure_months'].isna())
).astype(int)

#Считаю медиану по возрасту (только по корректным)
age_median = (
    df[df['tenure_months'] >= 0]
    .groupby('age_years')['tenure_months']
    .median()
)

# Заменяю отрицательные
mask_negative = df['tenure_months'] < 0
df.loc[mask_negative, 'tenure_months'] = (
    df.loc[mask_negative, 'age_years']
      .map(age_median)
)

#Считаю глобальную медиану для оставшихся NaN
global_median = df['tenure_months'].median()
df['tenure_months'] = df['tenure_months'].fillna(global_median)
```

```python
#Теперь пропусков нет, но есть нули, что нормально
print(df['tenure_months'].isna().sum())
print((df['tenure_months'] == 0).sum())
```

```python
df['tenure_negative_flag'] = (df['tenure_months'] < 0).astype(int)
df['tenure_months'] = df['tenure_months'].clip(lower=0)
```

Превращаю тип данных даты из строки в дату и разбиваю на колонки по годам и месяцам, чтобы суметь потсроить лаги и создать конечную необходимую матрицу сотрудник-признак для построения предсказаний.

```python
df['Дата начала'] = pd.to_datetime(df['Дата начала'])

df['year'] = df['Дата начала'].dt.year
df['month'] = df['Дата начала'].dt.month
```

Считаю границы дат , по которым у нас есть исторические данные и строю ранжировку по ВСЕМ месяцев, а не только по тем, когда были отпуски, чтобы предсказание было полным.

```python
min_date = df['Дата начала'].min()
max_date = df['Дата начала'].max()

all_months = pd.date_range(
    start=min_date,
    end=max_date,
    freq='MS'   # Month Start
)

calendar_df = pd.DataFrame({
    'year': all_months.year,
    'month': all_months.month
})
```

Беру уникальные айдишки сотрудников и создаю декартово произведение , чтобы получить вид (каждый сотрудник × каждый месяц), затем соединяю эти таблицы cross join методом, чтобы получить полный датасет.

```python
employees = df[['employee_id']].drop_duplicates()
```

```python
employees['key'] = 1
calendar_df['key'] = 1

employee_month = (
    employees
    .merge(calendar_df, on='key')
    .drop('key', axis=1)
)
```

```python
vacations = (
    df.groupby(['employee_id', 'year', 'month'], as_index=False)
      .agg(
          vacation_days=('Кол-во календарных дней', 'sum'),
          vacations_cnt=('Кол-во календарных дней', 'size'),
          start_min=('Дата начала', 'min')   # пригодится для регрессии даты начала
      )
)
```

```python
ml_df = employee_month.merge(
    vacations,
    on=['employee_id', 'year', 'month'],
    how='left'
)

ml_df['vacation_flag'] = ml_df['vacation_days'].notna().astype('int8')
ml_df['vacation_days'] = ml_df['vacation_days'].fillna(0).astype('int16')
ml_df['vacations_cnt'] = ml_df['vacations_cnt'].fillna(0).astype('int8')
```

Таким образом из таблицы, где была информация по отпускам, я получила таблицу информации по сотрудникам с начала периода сбора данных по КАЖДОМУ месяцу.

```python
ml_df.head()
```

```python
ml_df['is_summer_month'] = ml_df['month'].isin([6,7,8]).astype('int8')
```

Берём признаки сотрудника из df, но только из периода до 2025 (чтобы не подглядывать в тест); также я не стала сюда добавлять колонку по отрицательному стажу, так как он будет линейно кореллировать с 'tenure_invalid' и вызывать шум.

**Лаги:**
Для начала сортирую датасет по возрастанию дат, а затем добавляю shift, что смещает мои лаги на месяц во избежание утечки, и теперь мы считаем сколько раз был отпуск у сотрудника благодаряю лагам за ПРЕДЫДУЩИЙ период с окнами: 3 месяца, 6 месяцев.

```python
ml_df = ml_df.sort_values(['employee_id', 'year', 'month'])
#shift сдвигает месяцы и считает за ПРЕДЫДУЩИЕ флаги, чтоюы не было утечки.
ml_df['vac_flag_lag1'] = ml_df.groupby('employee_id')['vacation_flag'].shift(1).fillna(0).astype('int8')
ml_df['vac_days_lag1'] = ml_df.groupby('employee_id')['vacation_days'].shift(1).fillna(0).astype('int16')

ml_df['vac_flag_last3'] = (
    ml_df.groupby('employee_id')['vacation_flag']
         .shift(1)
         .rolling(3)
         .sum()
         .reset_index(level=0, drop=True)
         .fillna(0)
         .astype('int8')
)

ml_df['vac_flag_last6'] = (
    ml_df.groupby('employee_id')['vacation_flag']
         .shift(1)
         .rolling(6)
         .sum()
         .reset_index(level=0, drop=True)
         .fillna(0)
         .astype('int8')
)
```

Анализ целевой переменной показал, что в 2020–2021 годах данные об отпусках были существенно неполными (в 80 раз меньше зарегистрированных отпусков по сравнению с 2022 годом). В связи с этим данные за 2020–2021 были исключены из обучающей выборки

```python
#теперь данные распределяются лучше, ибо до этого в train у меня был target mean = 0,019.
train_df = ml_df[(ml_df['year'] >= 2022) & (ml_df['year'] <= 2023)].copy()
valid_df = ml_df[ml_df['year'] == 2024].copy()
test_df  = ml_df[ml_df['year'] == 2025].copy()

print("Train:", train_df.shape)
print("Valid:", valid_df.shape)
print("Test :", test_df.shape)

print("Target mean:")
print("Train:", train_df['vacation_flag'].mean())
print("Valid:", valid_df['vacation_flag'].mean())
print("Test :", test_df['vacation_flag'].mean())
```

Что касается заполнение числовых признаков, было бы глупо заполнить все медианой, поэтому я разделила их на части.

```python
X_train['birth_month'] = X_train['birth_month'].fillna(6)
X_train['north_bonus_days'] = X_train['north_bonus_days'].fillna(0)
```

```python
continuous_cols = [
    'department_size',
    'age_years',
    'tenure_months'
]
count_cols = [
    'children_count',
    'school_children_count',
    'extra_days_irregular'
]
```

```python
#Заполнение непрерывных признаков медианой (только по train)

medians = {}

for col in continuous_cols:
    if col in X_train.columns:
        med = X_train[col].median()
        medians[col] = med

        X_train[col] = X_train[col].fillna(med)
        X_valid[col] = X_valid[col].fillna(med)
        X_test[col]  = X_test[col].fillna(med)

#Заполнение счётчиков нулями

for col in count_cols:
    if col in X_train.columns:
        X_train[col] = X_train[col].fillna(0)
        X_valid[col] = X_valid[col].fillna(0)
        X_test[col]  = X_test[col].fillna(0)

#Проверка
print("NaN в X_train:", X_train[continuous_cols + count_cols].isna().sum().sum())
print("NaN в X_valid:", X_valid[continuous_cols + count_cols].isna().sum().sum())
print("NaN в X_test :", X_test[continuous_cols + count_cols].isna().sum().sum())
```
