```python
min_date = df['Дата начала'].min()
max_date = df['Дата начала'].max()

all_months = pd.date_range(
    start=min_date,
    end=max_date,
    freq='MS'   # Month Start
)

calendar_df = pd.DataFrame({
    'year': all_months.year,
    'month': all_months.month
})
```

```python
employees = df[['employee_id']].drop_duplicates()
```

```python
employees['key'] = 1
calendar_df['key'] = 1

employee_month = (
    employees
    .merge(calendar_df, on='key')
    .drop('key', axis=1)
)
```

```python
df['year'] = df['Дата начала'].dt.year
df['month'] = df['Дата начала'].dt.month

vacations = (
    df.groupby(['employee_id', 'year', 'month'], as_index=False)
      .agg(
          vacation_days=('Кол-во календарных дней', 'sum'),
          vacations_cnt=('Кол-во календарных дней', 'size'),
          start_min=('Дата начала', 'min')   # пригодится для регрессии даты начала
      )
)
```

```python
ml_df = employee_month.merge(
    vacations,
    on=['employee_id', 'year', 'month'],
    how='left'
)

ml_df['vacation_flag'] = ml_df['vacation_days'].notna().astype('int8')
ml_df['vacation_days'] = ml_df['vacation_days'].fillna(0).astype('int16')
ml_df['vacations_cnt'] = ml_df['vacations_cnt'].fillna(0).astype('int8')
```

```python
ml_df['is_summer_month'] = ml_df['month'].isin([6,7,8]).astype('int8')

```

```python
# берём признаки сотрудника из df, но только из периода до 2025 (чтобы не подглядывать в тест)
df_static_src = df[df['year'] <= 2024].sort_values('Дата начала')

static_cols = [
    'employee_id',
    'department', 'department_size', 'region',
    'is_manager', 'manager_avoids_summer',
    'is_north_region', 'north_bonus_days',
    'age_years', 'birth_month', 'is_married',
    'children_count', 'school_children_count', 'has_school_children',
    'has_irregular_schedule', 'extra_days_irregular',
    'days_available',
    'tenure_months', 'is_new_employee'
]

static_df = (
    df_static_src.groupby('employee_id', as_index=False)[static_cols[1:]].last()
)
```

```python
static_df['tenure_negative_flag'] = (static_df['tenure_months'] < 0).astype('int8')
static_df['tenure_months'] = static_df['tenure_months'].clip(lower=0)
```

```python
ml_df = ml_df.merge(static_df, on='employee_id', how='left')
```

**Лаги:**
отпуск в прошлом месяце

количество отпусков за 3 месяца

количество отпусков за 6 месяцев

дни отпуска в прошлом году

```python
ml_df = ml_df.sort_values(['employee_id', 'year', 'month'])
#shift сдвигает месяцы и считает за ПРЕДЫДУЩИЕ флаги, чтоюы не было утечки.
ml_df['vac_flag_lag1'] = ml_df.groupby('employee_id')['vacation_flag'].shift(1).fillna(0).astype('int8')
ml_df['vac_days_lag1'] = ml_df.groupby('employee_id')['vacation_days'].shift(1).fillna(0).astype('int16')

ml_df['vac_flag_last3'] = (
    ml_df.groupby('employee_id')['vacation_flag']
         .shift(1)
         .rolling(3)
         .sum()
         .reset_index(level=0, drop=True)
         .fillna(0)
         .astype('int8')
)

ml_df['vac_flag_last6'] = (
    ml_df.groupby('employee_id')['vacation_flag']
         .shift(1)
         .rolling(6)
         .sum()
         .reset_index(level=0, drop=True)
         .fillna(0)
         .astype('int8')
)
```

```python
train_df = ml_df[ml_df['year'] <= 2024].copy()
test_df  = ml_df[ml_df['year'] == 2025].copy()

print(train_df.shape, test_df.shape)
print(train_df['vacation_flag'].mean(), test_df['vacation_flag'].mean())
```
